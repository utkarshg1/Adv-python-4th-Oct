# Create a class person
1. attributes - properties of the class
2. methods - perfrom certain operations based on attributes

```python
class Person:

    # Constructor function __init__ to define attributes of class
    # self is any object of class person
    def __init__(self, name: str, age: int, gender: str, occupation: str):
        self.name = name
        self.age = age
        self.gender = gender
        self.occupation = occupation

    # this function is the string representaion of the class
    def __str__(self):
        return f"Person(name = {self.name}, age = {self.age}, gender = {self.gender}, occupation = {self.occupation})"

    # Write a method for class person
    def introduce(self):
        print(f"My name is {self.name}")
        print(f"My age is {self.age}")
        print(f"I am {self.gender}")
        print(f"I work as {self.occupation}")
```

```python
# p1 is object of class person
p1 = Person(name="Rahul", age=23, gender="male", occupation="Engg.")
print(type(p1))
```

# Dataclasses

```python
from dataclasses import dataclass
from typing import Literal

@dataclass
class Person2:
    name: str
    age: int
    gender: Literal["male", "female"]
    occuptation: str

    def intro(self):
        print(f"My name is {self.name}")
        print(f"My age is {self.age}")
        print(f"I am {self.gender}")
        print(f"I work as {self.occuptation}")
```

```python
p3 = Person2(name="Raman", age=25, gender="male", occuptation="HR")
print(p3)
```

# Bank account 


```python
@dataclass
class Account:
    ac_no: int
    name: str
    balance: float
    
    def check_balance(self):
        print(f"Account Number : {self.ac_no}, Name : {self.name}, Balance : {self.balance}")

    def deposit(self, amt: float):
        if amt <= 0:
            raise ValueError("The amount deposited cannot be negative or zero")
        self.balance = self.balance + amt

    def withdraw(self, amt: float):
        if amt <= 0:
            raise ValueError("The amount withdrawn cannot be negative or zero")
        if amt > self.balance:
            raise ValueError("Insufficient balance")
        self.balance = self.balance - amt

    def transfer(self, amt: float, other_ac: 'Account'):
        self.withdraw(amt)
        other_ac.deposit(amt)
```

```python
ac1 = Account(ac_no=1234, name="Raman", balance=0)
print(ac1)
```

```python
ac1.check_balance()
```

```python
ac1.deposit(1000.0)
```

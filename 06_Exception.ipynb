# With Exception handling

Any error is stored in Exception Class in python

```python
def division2(num: int | float, den: int | float) -> float:
    try:
        return num / den
    except Exception as e:
        print(f"Exception Occured : {e}")
```

```python
d1 = division2(num=20, den=10)
print(d1)
print("Done")
```

```python
d2 = division2(25, 0)
print(d2)
print("Done")
```

```python
d3 = division2("Three", 4)
print(d3)
print("Done")
```

# Custom exception handling
1. ZeroDivisionError
2. TypeError

```python
def division3(num: int | float, den: int | float) -> float:
    try:
        return num / den
    except ZeroDivisionError:
        print("Denominator cannot be 0 in division")
    except TypeError:
        print("Numerator and denominator should be a number")
```

```python
r1 = division3(100, 20)
print(r1)
print("Done")
```

```python
r2 = division3(50, 3)
print(r2)
print("Done")
```

```python
r3 = division3(100, 0)
print(r3)
print("Done")
```

```python
r4 = division3("Five", 2)
print(r4)
print("Done")
```

```python
r5 = division3("a", "b")
print(r5)
print("Done")
```

# Raise custom errors
I want to take age as input from user

age should be between 0 to 100

```python
def age_validator():
    age = int(input("Please enter age in years : "))
    if age <= 0 or age > 100:
        raise ValueError("Age must be between 1 to 100")
    return age
```

```python
# Provide age 25
age1 = age_validator()
print(age1)
```

```python
# Negative age : -4
age2 = age_validator()
print(age2)
```

```python
# Age value 200
age3 = age_validator()
print(age3)
```

```python
# With try except block
try:
    age4 = age_validator()
    print(f"Valid age entered : {age4}")
except Exception as e:
    print(f"Exception Occured : {e}")
```
